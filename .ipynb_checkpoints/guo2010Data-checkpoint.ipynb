```python
raw_df = pd.read_csv("./data/guo2010Data.txt", header='infer', index_col=0, sep="\t", encoding=sys.getfilesystemencoding())
raw_df = raw_df.transpose()

cell_stage = raw_df['Stage'].astype('int')
raw_df = raw_df.drop('Stage',1)

cell_names = raw_df.index.values
gene_names = raw_df.columns.values

raw_df.shape[0]
```

```python
from SCOUT import *
```

```python
Z = SCOUT(raw_df, stages=cell_stage)
V = Z.dim_down(method='lle',rand_seed=6)
Z.plotD(V)
```

```python
centroids,labels = Z.clustering(V,method='gmm',nclust=15)
# Tcsr,edges = Z.get_MST()
Z.plotD(V)
```

## Wishbone

```python
raw_df.to_csv('guo2010Data.csv',index=True,header=True)
pd.DataFrame(raw_df)
```

```python
import wishbone
import os
scdata = wishbone.wb.SCData.from_csv(os.path.expanduser('guo2010Data.csv'), data_type='sc-seq', normalize=True)
scdata
```

```python
scdata.run_pca()
```

```python
scdata.run_tsne(n_components=2, perplexity=30)
scdata.run_diffusion_map()
scdata.run_diffusion_map_correlations()
wb = wishbone.wb.Wishbone(scdata)
```

```python
wb.run_wishbone(start_cell='1C 1',k=120,branch=True,components_list=[1, 2], num_waypoints=200)
```

```python
cell_scores = wb.trajectory
order = sorted(range(len(cell_scores)), key=lambda k: cell_scores[k])
order = [cell_stage.values[i] for i in order]
print(order)
```

```python
from scipy import stats
cor = stats.pearsonr(order,cell_stage.values)[0]
print(cor)
```
